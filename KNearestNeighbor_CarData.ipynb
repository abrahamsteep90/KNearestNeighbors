```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model,preprocessing
```

```python
data = pd.read_csv('car.data')
```

```python
data.head()
```

```python
data.rename(columns={"vhigh": "buying", "vhigh.1": "maint", "2": "doors", "2.1": "persons", "small": "lug_boot", "low": "safety", "unacc": "class"}, inplace= True)
data.head()
```

```python
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"]))
maint= le.fit_transform(list(data["maint"]))
doors = le.fit_transform(list(data["doors"]))
persons = le.fit_transform(list(data["persons"]))
lug_boot = le.fit_transform(list(data["lug_boot"]))
safety = le.fit_transform(list(data["safety"]))
classValues = le.fit_transform(list(data["class"]))
```

```python
predict = "class"
X = list(zip(buying, maint, doors, persons, lug_boot, safety))
y = list(classValues)
```

```python
x_train, x_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size =0.2)
```

```python
model = KNeighborsClassifier(n_neighbors = 7)
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
accuracy
```

```python
predictions = model.predict(x_test)
```

```python
names = ["unacc", "acc", "good", "vgood" ]
for x in range(len(predictions)):
    print( "Predictions :", predictions[x], "Data:", x_test[x], "Actual:", y_test[x], names[y_test[x]])
    distance = model.kneighbors([x_test[x]], 7, True)
    print("distance:", distance)
```
